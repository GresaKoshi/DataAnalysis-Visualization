```python
def calculate_smth(*args):

  if not args:
    return 0

  first_number = args[0]
  product_of_others = 1
  for i in range(1, len(args)):
    product_of_others *= args[i]

  number_of_parameters = len(args)

  if number_of_parameters == 0:
    return 0

  return (first_number + product_of_others) / number_of_parameters


result = calculate_smth(1, 2, 4)
print(result)
```

```python
def format_dates_without_datetime(date_strings):

    formatted_dates = []
    month_names = [
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ]
    for date_str in date_strings:
        try:
            day_str, month_str, year_str = date_str.split('/')
            day = int(day_str)
            month = int(month_str)
            year = int(year_str)

            if 1 <= month <= 12 and 1 <= day <= 31:  # Basic date validation
                month_name = month_names[month - 1]
                formatted_date_str = f"{month_name} {day:02d}, {year}"
                formatted_dates.append(formatted_date_str)
            else:
                print(f"Error: Invalid date '{date_str}'. Skipping.")
        except ValueError:
            print(f"Error: Could not parse date string '{date_str}'. Skipping.")
    return formatted_dates

dates = ['12/02/2025', '15/03/2025', '17/04/2025']
formatted_list = format_dates_without_datetime(dates)
print(formatted_list)
```

```python
import numpy as np
arr = np.array([
    [1.0, 2.0, np.nan, 4.0],
    [5.0, np.nan, 7.0, 8.0],
    [9.0, 10.0, 11.0, np.nan]
])
```

```python
row_means = np.nanmean(arr, axis=1, keepdims=True)
nan_mask = np.isnan(arr)
filled_arr = np.where(nan_mask, row_means, arr)

print(filled_arr)
```

```python
greater_than_5_mask = filled_arr > 5
values_greater_than_5 = filled_arr[greater_than_5_mask]

print("\nValues greater than 5:")
print(values_greater_than_5)

```

```python
column_maximum = np.nanmax(arr, axis=0)

print("\nMaximum value for each column :")
print(column_maximum)
```

```python
df = pd.read_csv('hr_data.csv')

def categorize_experience(years):

    if years < 2:
        return 'Junior'
    elif 2 <= years <= 5:
        return 'Mid'
    else:
        return 'Senior'

```

```python
df['ExperienceCategory'] = df['YearsAtCompany'].apply(categorize_experience)

print(df[['EmpID', 'YearsAtCompany', 'ExperienceCategory']])
```

```python
def extract_numeric_id(emp_id):
    try:
        return int(emp_id.replace('EMP', ''))
    except AttributeError:
        return None
```

```python
df['NumericID'] = df['EmpID'].apply(extract_numeric_id)
```

```python
df.set_index('NumericID', inplace=True)
```

```python
average_salary_by_performance = df.groupby('PerformanceScore')['Salary'].mean()

print("Average Salary by PerformanceScore:")
print(average_salary_by_performance)
```

```python
average_salary_by_years = df.groupby('YearsAtCompany')['Salary'].mean()

print("Average Salary by YearsAtCompany:")
print(average_salary_by_years)
```

```python
corr_years_salary = df['YearsAtCompany'].corr(df['Salary'])
print(f"YearsAtCompany - Salary Correlation: {corr_years_salary:.2f}")

performance_order = {'Poor': 1, 'Good': 2, 'Average': 3, 'Excellent': 4}
df['PerformanceScore_Numeric'] = df['PerformanceScore'].map(performance_order)
corr_performance_salary = df['PerformanceScore_Numeric'].corr(df['Salary'])
print(f"PerformanceScore - Salary Correlation: {corr_performance_salary:.2f}")
```
